```
*Ejercicio 1*
Seleccionar las canciones de Iron Maiden

SELECT artists.Name, albums.AlbumId,tracks.Name  
FROM tracks A, albums B, artists C
where
B.AlbumId = B.AlbumId and
B.ArtistId = C.ArtistId and
C.Name = 'Iron Maiden'
 

*Ejercicio 2*
Discos con más de 25 canciones

SELECT A.Title, COUNT(B.Name) as Q_canciones_disco from albums A, tracks B
where  A.AlbumId = B.AlbumId 
GROUP BY A.AlbumId
HAVING Q_canciones_disco > 25
ORDER BY Q_canciones_disco 

*Ejercicio 3*
Las 10 canciones más populares

select Tr.Name, count(B.trackid) as Qtracks from tracks A, playlist_track B
where A.TrackId = B.TrackId
GROUP By A.Name
ORDER by  Qtracks DESC
LIMIT 10
```
